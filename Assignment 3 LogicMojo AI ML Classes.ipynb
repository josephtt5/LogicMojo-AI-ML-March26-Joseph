Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()

features = iris.data

mean=np.mean(features,axis=0)
median=np.median(features,axis=0)
standard=np.std(features,axis=0)
variance = np.var(features,axis=0)

for i, name in enumerate(iris.feature_names):
    print(f"--- {name} ---")
    print(f"Mean: {mean[i]} | Median: {median[i]}")
    print(f"Std Dev: {standard[i]} | Variance: {variance[i]}\n")

#If one feature has a range of 1–7 (high variance) and another is 0.1–0.5 (low variance),
#the model will think the high-variance feature is "more important" simply because the numbers are bigger.

sepal_length= iris.data[:,0].reshape(-1,1)
sepal_width= iris.data[:,1].reshape(-1,1)
petal_length= iris.data[:,2].reshape(-1,1)
petal_width= iris.data[:,3].reshape(-1,1)


print(sepal_length)
print(sepal_width)
print(petal_length)
print(petal_width)

#Models expect this format for uniformity and matric multiplication 
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
from sklearn.datasets import load_iris
import numpy as np

iris= load_iris()
dataset= iris.data

corr_matric = np.corrcoef(dataset,rowvar=False)

print(corr_matric)
#Higly correlated features 1-3 1-4 3-4 
#We drop column 4 as it is identical with 3 and will create unesecary weight on the model

data_new = np.delete(dataset,3,axis=1)

corr_matric = np.corrcoef(data_new,rowvar=False)

print(corr_matric)

#overfitting is when the model is so complex that it memoroizes noise
#it is better to have lower accuracy with less features that a bit higher accuracy with a lot more features
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns

df = sns.load_dataset('titanic')

def age_group(age):
    if age > 50 :
        return 'Senior'
    elif age > 18:
        return 'Adult'
    else:
        return 'Child'
        
df['age_group']= df['age'].apply(age_group)  
print(df)

print(f"Survival by age group: {df.groupby('age_group')['survived'].mean().map('{:.2%}'.format)}")
print(f"Survival by age group: {df.groupby(['age_group','sex'])['survived'].mean().map('{:.2%}'.format)}")

#Adults has the highest survival rate
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

np.random.seed(42)
errors = np.random.normal(0,1,1000)

mean= np.mean(errors)
std = np.std(errors)

outliers = errors[(errors< (mean - 2 * std)) | ( errors > (mean + 2 *std))]

print(outliers)

#High variance is when a o model with simila dataset produce complute different results
#this is due to complex modeling,, this create incostencies.
```
